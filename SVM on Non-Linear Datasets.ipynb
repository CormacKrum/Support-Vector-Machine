```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.model_selection import cross_val_score
from mpl_toolkits.mplot3d import Axes3D
```

```python
X,y = make_circles(n_samples=500,noise=0.02)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
x1 = X[:,0]
x2 = X[:,1]
x3 = x1**2 + x2**2
X_ = np.zeros((X.shape[0],3))
X_[:,:-1] = X
X_[:,-1] = x3
X=X_
```

```python
def plot3D(X,y):
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111,projection='3d')
    x1 = X[:,0]
    x2 = X[:,1]
    x3 = X[:,2]
    ax.scatter(x1,x2,x3,zdir='z',s=20,c=y,depthshade=True)
    return ax
```

```python
ax = plot3D(X,y)
plt.show()
```

# Kernel Based Classifcation in SVM's

Sklearn supports the following types of Kernels, which can be used in many-real life problems.

* Linear Kernel
* RBF Kernel
* Polynomial Kernel
* Sigmoid Kernel

Kernel trick is method of using a linear classifer to solve a non-linear problem. It transforms the linearly in-separable data into linearly separable one.

```python
from sklearn.svm import SVC
```

```python
modelL = SVC(kernel='linear')
modelL.fit(X[:,:-1],y)
print(modelL.score(X[:,:-1],y))
```

```python
modelR = SVC(kernel='rbf')
modelR.fit(X[:,:-1],y)
print(modelR.score(X[:,:-1],y))
```

```python
modelP = SVC(kernel='poly')
modelP.fit(X[:,:-1],y)
print(modelP.score(X[:,:-1],y))
```

```python
modelS = SVC(kernel='sigmoid')
modelS.fit(X[:,:-1],y)
print(modelS.score(X[:,:-1],y))
```

```python
def custom_kernel(x1,x2):
    return np.square(np.dot(x1,x2.T))
```

```python
modelC = SVC(kernel=custom_kernel)
modelC.fit(X[:,:-1],y)
print(modelC.score(X[:,:-1],y))
```
